```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y =make_moons(n_samples=1000,noise=0.2)
```

```python
df = {'x1':x[:,0],'x2':x[:,1],'y':y}
```

```python
dataset= pd.DataFrame(df)
```

```python
dataset
```

```python
sns.scatterplot(x=dataset["x1"],y=dataset["x2"],hue=dataset["y"])
```

```python
x_a = dataset[["x1","x2"]]
x_a
```

```python
y_a = dataset["y"]
y_a
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_a,y_a,train_size=0.3,random_state=43)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
# bagging beta estimator
bg = BaggingClassifier(estimator=SVC(),n_estimators=30)# n_estimators = n0 0f models
bg.fit(x_train,y_train)
bg.score(x_train,y_train)*100, bg.score(x_test,y_test)*100
```

```python
rf = RandomForestClassifier(n_estimators=90)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)*100 , rf.score(x_test,y_test)*100
```
